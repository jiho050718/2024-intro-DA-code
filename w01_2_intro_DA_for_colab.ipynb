```python
import pandas as pd

pop = pd.read_csv("202408_202408_주민등록인구및세대현황_월간.csv", encoding='cp949', index_col=0, thousands=',')
pop
```

```python
pop.columns = ['총 인구수', '세대수', '세대당인구', '남자인구수', '여자인구수', '남여비율']
pop
```

```python
pop.index = [pop.index[i].split()[0] for in range(len(pop))]
pop
```

```python
import matplotlib.pyplot as pit
import seaborn as sns

sns.scatterplot(pop['총 인구수'])
pit.xticks(rotation=90)
pit.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

pop1 = pop.sort_values('총 인구수')
sns.barplot(x=pop1.index, y='총 인구수', data=pop1)
pit.xticks(rotation=90)
pit.show()
```
